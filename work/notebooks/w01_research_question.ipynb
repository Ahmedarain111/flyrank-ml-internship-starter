## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Decision:** The work supports the decision of which pages should be reviewed first by a content or SEO team.

**Action:** Reviewers can inspect the highest-ranked pages and decide whether they should be refreshed, expanded, monitored, or otherwise improved.

**Cost of a wrong recommendation:** A false positive wastes reviewer time on pages that do not need attention. A false negative means a page that genuinely needs review may continue losing visibility or engagement. Because of these costs, the model should be treated as decision support rather than an automatic decision-maker.
